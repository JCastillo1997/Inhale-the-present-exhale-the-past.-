```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt_2
import pylab as plt
```

```python
df=pd.read_csv('calidad_aire.csv')
df.head()
```

```python
df_madrid =df[df['CCAA']=='Comunidad de Madrid']
plt.title('Calidad del aire agrupada Madrid')
sns.histplot(data=df_madrid, x= 'PORCENTAJE_CALIDAD_AIRE')
plt.xticks(rotation=45, ha='right')
```
