```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt_2
import pylab as plt
```

```python
df=pd.read_csv('calidad_aire.csv')
df.head()
```

```python
df_madrid =df[df['CCAA']=='Comunidad de Madrid']
sns.histplot(data=df_madrid, x= 'PORCENTAJE_CALIDAD_AIRE',palette="dark")
```

```python
plt.figure(figsize=(7,5))
plt.title('CCAA y Aire')
sns.barplot(df, x="CCAA", y="PORCENTAJE_CALIDAD_AIRE",palette="muted")
plt.xticks(rotation=45,ha='right')
plt.xlabel('Comunidad Autonoma')
plt.ylabel('Calidad del Aire')
plt.show()
```
